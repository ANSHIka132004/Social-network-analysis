```python
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return x + y

x0, y0 = 0, 1
h = 0.1
n = 10
x = np.zeros(n + 1)
y = np.zeros(n + 1)
x[0], y[0] = x0, y0

for i in range(n):
    y[i + 1] = y[i] + h * f(x[i], y[i])
    x[i + 1] = x[i] + h
print("x-values:", x)
print("y-values:", y)

plt.plot(x, y, 'o-', label="Euler Approximation")
plt.title("Eulerâ€™s Method for dy/dx = x + y")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def f(x, y):
    return -2 * y  # dy/dx = -2y

x0, y0, h, n = 0, 1, 0.1, 10
x, y = [x0], [y0]

for i in range(n):
    y_next = y[i] + h * f(x[i], y[i])
    x_next = x[i] + h
    x.append(x_next)
    y.append(y_next)

print("x-values:", [round(val, 2) for val in x])
print("y-values:", [round(val, 6) for val in y])
```

```python
def f(x, y):
    return y - x**2 + 1  # dy/dx = y - x^2 + 1

x0, y0, h, n = 0, 0.5, 0.1, 10
x, y = [x0], [y0]

for i in range(n):
    y_next = y[i] + h * f(x[i], y[i])
    x_next = x[i] + h
    x.append(x_next)
    y.append(y_next)

print("x-values:", [round(val, 2) for val in x])
print("y-values:", [round(val, 6) for val in y])
```
